## Customers `Churn` Mapping 

```python
# mapping yes and no value these colums Partner,Dependents,PhoneService
yes_no_mapping = {'Yes':1,'No':0}
Churn_data['Partner'] = Churn_data['Partner'].map(yes_no_mapping)
Churn_data['Dependents'] = Churn_data['Dependents'].map(yes_no_mapping)
Churn_data['PhoneService'] = Churn_data['PhoneService'].map(yes_no_mapping)
```

```python
Churn_data.head()
```

```python
#get columns in data set
Churn_data.columns
```

```python
# mapping yes and no value these colums OnlineSecurity,OnlineBackup,DeviceProtection,TechSupport,StreamingMovies,StreamingTV
internet_service_mapping = {'Yes':1,'No':0,'No internet service':0}
Churn_data['OnlineSecurity'] = Churn_data['OnlineSecurity'].map(internet_service_mapping)
Churn_data['OnlineBackup'] = Churn_data['OnlineBackup'].map(internet_service_mapping)
Churn_data['DeviceProtection'] = Churn_data['DeviceProtection'].map(internet_service_mapping)
Churn_data['TechSupport'] = Churn_data['TechSupport'].map(internet_service_mapping)
Churn_data['StreamingMovies'] = Churn_data['StreamingMovies'].map(internet_service_mapping)
Churn_data['StreamingTV'] = Churn_data['StreamingTV'].map(internet_service_mapping)
```

```python
# mapping yes and no value these colums MultipleLines
phone_service_maping = {'Yes':1,'No':0,'No phone service' :0}
Churn_data['MultipleLines'] = Churn_data['MultipleLines'].map(phone_service_maping)
```

```python
Churn_data
```

```python
# check values in InternetService
Churn_data['InternetService'].value_counts()
```

```python
# mapping DSL,No, and Fiber optic value these colums InternetService
DSLFIBER_mapping ={'DSL':1, 'Fiber optic':1,'No':0}
Churn_data['InternetService'] = Churn_data['InternetService'].map(DSLFIBER_mapping)
```

```python
Churn_data
```

```python
# mapping Month-to-month,One year, and Two year  these colums Contract
contract_mapping = {'Month-to-month':'MM','One year':'OY','Two year':'TY'}
Churn_data['Contract'] = Churn_data['Contract'].map(contract_mapping)
```

```python
Churn_data
```

```python
# mapping  PaperlessBilling 
Churn_data['PaperlessBilling'] = Churn_data['PaperlessBilling'].map(yes_no_mapping)
```

```python
Churn_data
```

```python
#get value count in PaymentMethod
Churn_data['PaymentMethod'].value_counts()
```

```python
# mapping  PaymentMethod 
payment_mapping = {'Bank transfer (automatic)':'Bank','Credit card (automatic)':'Card','Electronic check':'Check','Mailed check':'Check'}
Churn_data['PaymentMethod'] = Churn_data['PaymentMethod'].map(payment_mapping)
```

```python
Churn_data
```

```python
Churn_data['Churn'] = Churn_data['Churn'].map(yes_no_mapping)
```
