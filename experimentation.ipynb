### Train Test Split

```python
from sklearn.model_selection import train_test_split
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
# y = mX + c
```

```python
X = Churn_data[["customerID", "gender", "SeniorCitizen", "Partner", "Dependents", "tenure"]]
```

```python
type(X)
```

```python
X.head()
```

```python
y = Churn_data["tenure"]
```

```python
type(y)
```

```python
y[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
train_data, test_data = train_test_split(Churn_data, test_size=0.33, random_state=42)
```

```python
train_data.shape, test_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
label = 'tenure'
print("Summary of class variable: \n", train_data[label].describe())
```

```python
save_path = 'models'
predictor = TabularPredictor(label=label, path=save_path).fit(train_data)
```

```python
predictor.leaderboard(silent=True)
```

## Model Evaluation

```python
test_data.head()
```

```python
y_test = test_data[label]  # values to predict
```

```python
y_test[0:5]
```

```python
test_data_nolab = test_data.drop(columns=[label])  # delete label column to prove we're not cheating
test_data_nolab.head()
```

```python
save_path
```

Now, load in the stored models.

```python
predictor
```

```python
save_model_predictor = TabularPredictor.load(save_path)
```

```python
save_model_predictor
```

```python
y_pred = save_model_predictor.predict(test_data_nolab)
```

```python
y_pred[0:5]
```

```python
print("Predictions:  \n", y_pred)
perf = save_model_predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
save_model_predictor.leaderboard(test_data, silent=True)
```

## Bringing it all together
